```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the Data
X1 = np.array([1500, 1800, 1200, 2000, 1700, 1600, 1900, 1100, 1400, 2100])  # Size
X2 = np.array([3, 4, 2, 4, 3, 3, 4, 2, 3, 5])  # Bedrooms
Y = np.array([300000, 350000, 250000, 400000, 320000, 310000, 360000, 230000, 290000, 420000])  # Price

# Number of data points
n = len(X1)

# Step 2: Compute Summations
sum_X1 = np.sum(X1)
sum_X2 = np.sum(X2)
sum_Y = np.sum(Y)
sum_X1_sq = np.sum(X1**2)
sum_X2_sq = np.sum(X2**2)
sum_X1Y = np.sum(X1 * Y)
sum_X2Y = np.sum(X2 * Y)
sum_X1X2 = np.sum(X1 * X2)

# Step 3: Solve for Coefficients using Normal Equations
A = np.array([
    [n, sum_X1, sum_X2],
    [sum_X1, sum_X1_sq, sum_X1X2],
    [sum_X2, sum_X1X2, sum_X2_sq]
])

B = np.array([sum_Y, sum_X1Y, sum_X2Y])

# Solve for b0, b1, b2
B_solution = np.linalg.solve(A, B)
b0, b1, b2 = B_solution

# Step 4: Print Regression Equation
print(f"Regression Equation: Price = {b0:.2f} + {b1:.2f} * Size + {b2:.2f} * Bedrooms")

# Step 5: Generate Predictions
predicted_Y = b0 + b1 * X1 + b2 * X2

# Step 6: Visualization
plt.figure(figsize=(12, 5))

# Plot Price vs. Size
plt.subplot(1, 2, 1)
plt.scatter(X1, Y, color='blue', label="Actual Prices")
plt.plot(X1, predicted_Y, color='red', linestyle='dashed', label="Predicted Prices")
plt.xlabel("Size (sq. ft.)")
plt.ylabel("Price ($)")
plt.title("House Price vs. Size")
plt.legend()

# Plot Price vs. Bedrooms
plt.subplot(1, 2, 2)
plt.scatter(X2, Y, color='blue', label="Actual Prices")
plt.plot(X2, predicted_Y, color='red', linestyle='dashed', label="Predicted Prices")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price ($)")
plt.title("House Price vs. Bedrooms")
plt.legend()

plt.tight_layout()
plt.show()

# Step 7: Predict Price for New Inputs
def predict_price(size, bedrooms):
    price = b0 + b1 * size + b2 * bedrooms
    return price

# User Input for Prediction
size_input = float(input("Enter house size (sq. ft.): "))
bedrooms_input = int(input("Enter number of bedrooms: "))

predicted_price = predict_price(size_input, bedrooms_input)
print(f"Predicted Price for {size_input} sq. ft. and {bedrooms_input} bedrooms: ${predicted_price:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Read Data from CSV
df = pd.read_csv("house_data.csv")

# Extract columns
X1 = df["Size"].values
X2 = df["Bedrooms"].values
Y = df["Price"].values

# Number of data points
n = len(X1)

# Step 2: Compute Summations
sum_X1 = np.sum(X1)
sum_X2 = np.sum(X2)
sum_Y = np.sum(Y)
sum_X1_sq = np.sum(X1**2)
sum_X2_sq = np.sum(X2**2)
sum_X1Y = np.sum(X1 * Y)
sum_X2Y = np.sum(X2 * Y)
sum_X1X2 = np.sum(X1 * X2)

# Step 3: Solve for Coefficients using Normal Equations
A = np.array([
    [n, sum_X1, sum_X2],
    [sum_X1, sum_X1_sq, sum_X1X2],
    [sum_X2, sum_X1X2, sum_X2_sq]
])

B = np.array([sum_Y, sum_X1Y, sum_X2Y])

# Solve for b0, b1, b2
B_solution = np.linalg.solve(A, B)
b0, b1, b2 = B_solution

# Step 4: Print Regression Equation
print(f"Regression Equation: Price = {b0:.2f} + {b1:.2f} * Size + {b2:.2f} * Bedrooms")

# Step 5: Generate Predictions
predicted_Y = b0 + b1 * X1 + b2 * X2

# Step 6: 3D Visualization
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter Plot of Actual Data
ax.scatter(X1, X2, Y, color='blue', label="Actual Prices")

# Create a Grid for Predictions
X1_range = np.linspace(min(X1), max(X1), 20)
X2_range = np.linspace(min(X2), max(X2), 20)
X1_grid, X2_grid = np.meshgrid(X1_range, X2_range)
Y_pred_grid = b0 + b1 * X1_grid + b2 * X2_grid

# Surface Plot for Regression Plane
ax.plot_surface(X1_grid, X2_grid, Y_pred_grid, color='red', alpha=0.5)

# Labels and Titles
ax.set_xlabel("Size (sq. ft.)")
ax.set_ylabel("Bedrooms")
ax.set_zlabel("Price ($)")
ax.set_title("House Price Prediction (3D)")

plt.legend()
plt.show()

# Step 7: Predict Price for New Inputs
def predict_price(size, bedrooms):
    price = b0 + b1 * size + b2 * bedrooms
    return price

# User Input for Prediction
size_input = float(input("Enter house size (sq. ft.): "))
bedrooms_input = int(input("Enter number of bedrooms: "))

predicted_price = predict_price(size_input, bedrooms_input)
print(f"Predicted Price for {size_input} sq. ft. and {bedrooms_input} bedrooms: ${predicted_price:.2f}")

```
