```python
def is_prime(num):
    if num<2:
        return False
    for i in range(2,num//2+1):
        if num%i==0:
            return False
    return True
def numfrom(m,n):
    for num in range(m,n+1):
        if is_prime(num):
            print(num,end=" ")
numfrom(10,50)            
```

```python
def prints(m,n):
    if m==n:
        return n
    print(m)    
    prints(m+1,n)
prints(5,10)    
    
```
