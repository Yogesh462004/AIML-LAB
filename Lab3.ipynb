```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Years of Experience
Y = np.array([30000, 35000, 40000, 45000, 50000])  # Salary

# Step 1: Calculate mean of X and Y (ensure they are scalars)
X_mean = np.mean(X.flatten())
Y_mean = np.mean(Y.flatten())

# Step 2: Compute the slope (m)
numerator = np.sum((X.flatten() - X_mean) * (Y.flatten() - Y_mean))
denominator = np.sum((X.flatten() - X_mean) ** 2)

m = numerator / denominator  # Now should be correct

# Step 3: Compute intercept (b)
b = Y_mean - (m * X_mean)

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
print(f"Equation: Y = {m:.2f} X + {b:.2f}")

# Using sklearn's LinearRegression for verification
model = LinearRegression()
model.fit(X, Y)
print(20*'*')
print(f"Manual Slope: {m}")
print(f"manual Intercept:{b}")

plt.scatter(X,Y,color='blue',label='Actual Data')
plt.plot(X,model.predict(X),color='red',linewidth=2,label='Regression Line')
plt.ylabel('Salary')
plt.xlabel('Years of Experience')
plt.title('Simple linear Regression')
plt.legend()
plt.show()

##-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
print('-----------------------------------------------------------------------------------------------')
print(f"Sklearn Slope: {model.coef_[0]}")
print(f"Sklearn Intercept: {model.intercept_}")



plt.scatter(X,Y,color='blue',label='Actual Data')
plt.plot(X,model.predict(X),color='red',linewidth=2,label='Regression Line')
plt.ylabel('Salary')
plt.xlabel('Years of Experience')
plt.title('Simple linear Regression')
plt.legend()
plt.show()
def predict_sal(years):
    mp=m*years+b
    sp=model.predict(np.array([[years]]))[0]
    return mp,sp
years_of_experience=float(input('Enter years of experience:'))
manual_sal,sklearn_sal=predict_sal(years_of_experience)
print(f"\nPredicted Salary")
print(f"Manual Salary:${manual_sal:.2f}")
print(f"Scikit-Learn Prediction Salary:${sklearn_sal:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Years of Experience
Y = np.array([30000, 35000, 40000, 45000, 50000])  # Salary

model=DecisionTreeRegressor()
model.fit(X,Y)
years_of_experience=float(input('Enter years of experience:'))
predicted_salary=model.predict([[years_of_experience]])

print(f"Predicted Salary:${predicted_salary[0]:.2f}")

X_grid=np.arange(min(X),max(X),0.01).reshape(-1,1)
plt.scatter(X,Y,color='blue',label='Actual Data')
plt.plot(X_grid,model.predict(X_grid),color='red',linewidth=2,label='Decision Tree Prediction')
plt.scatter(years_of_experience,predicted_salary,color='green',s=100,label='Predicted Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Decisison Tree Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Step 1: Define the Data
X = np.array([
    [1500, 3], [1800, 4], [1200, 2], [2000, 4], [1700, 3],
    [1600, 3], [1900, 4], [1100, 2], [1400, 3], [2100, 5]
])  # Features: Size & Bedrooms

Y = np.array([300000, 350000, 250000, 400000, 320000, 310000, 360000, 230000, 290000, 420000])  # Price

# Step 2: Train the Decision Tree Model
decision_tree = DecisionTreeRegressor(random_state=0)
decision_tree.fit(X, Y)

# Step 3: Predict Price for New Inputs
def predict_price(size, bedrooms):
    return decision_tree.predict([[size, bedrooms]])[0]

# User Input for Prediction
size_input = float(input("Enter house size (sq. ft.): "))
bedrooms_input = int(input("Enter number of bedrooms: "))

predicted_price = predict_price(size_input, bedrooms_input)
print(f"Predicted Price for {size_input} sq. ft. and {bedrooms_input} bedrooms: ${predicted_price:.2f}")

# Step 4: Visualizing the Decision Tree Predictions
size_range = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
bedroom_range = np.linspace(min(X[:, 1]), max(X[:, 1]), 100)
predicted_prices = [predict_price(s, 3) for s in size_range]  # Keeping bedrooms fixed at 3

plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], Y, color='blue', label="Actual Prices")
plt.plot(size_range, predicted_prices, color='red', linestyle='dashed', label="Predicted Prices")
plt.xlabel("Size (sq. ft.)")
plt.ylabel("Price ($)")
plt.title("Decision Tree Regression: House Price vs. Size")
plt.legend()
plt.show()
```
