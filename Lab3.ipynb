```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Years of Experience
Y = np.array([30000, 35000, 40000, 45000, 50000])  # Salary

# Step 1: Calculate mean of X and Y (ensure they are scalars)
X_mean = np.mean(X.flatten())
Y_mean = np.mean(Y.flatten())

# Step 2: Compute the slope (m)
numerator = np.sum((X.flatten() - X_mean) * (Y.flatten() - Y_mean))
denominator = np.sum((X.flatten() - X_mean) ** 2)

m = numerator / denominator  # Now should be correct

# Step 3: Compute intercept (b)
b = Y_mean - (m * X_mean)

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
print(f"Equation: Y = {m:.2f} X + {b:.2f}")

# Using sklearn's LinearRegression for verification
model = LinearRegression()
model.fit(X, Y)
print(20*'*')
print(f"Manual Slope: {m}")
print(f"manual Intercept:{b}")

plt.scatter(X,Y,color='blue',label='Actual Data')
plt.plot(X,model.predict(X),color='red',linewidth=2,label='Regression Line')
plt.ylabel('Salary')
plt.xlabel('Years of Experience')
plt.title('Simple linear Regression')
plt.legend()
plt.show()

##-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
print('-----------------------------------------------------------------------------------------------')
print(f"Sklearn Slope: {model.coef_[0]}")
print(f"Sklearn Intercept: {model.intercept_}")



plt.scatter(X,Y,color='blue',label='Actual Data')
plt.plot(X,model.predict(X),color='red',linewidth=2,label='Regression Line')
plt.ylabel('Salary')
plt.xlabel('Years of Experience')
plt.title('Simple linear Regression')
plt.legend()
plt.show()
def predict_sal(years):
    mp=m*years+b
    sp=model.predict(np.array([[years]]))[0]
    return mp,sp
years_of_experience=float(input('Enter years of experience:'))
manual_sal,sklearn_sal=predict_sal(years_of_experience)
print(f"\nPredicted Salary")
print(f"Manual Salary:${manual_sal:.2f}")
print(f"Scikit-Learn Prediction Salary:${sklearn_sal:.2f}")
```
