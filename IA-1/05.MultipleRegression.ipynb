## MULTIPLE REGRESSION

<h3>Using Library</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Dataset
data = {
    'Experience': [1, 2, 3, 4, 5],
    'Education': [3, 4, 5, 6, 7],
    'Salary': [30000, 40000, 45000, 50000, 67000]
}
df = pd.DataFrame(data)

# Features and target
X = df[['Experience', 'Education']]
y = df['Salary']

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Output results
print("\nRegression")
print(f"Intercept: {model.intercept_:.2f}")
print("Coefficients:", [f"{c:.2f}" for c in model.coef_])

# Prediction for 3 years of Experience and Education level 5
experience = 3
education = 5
predicted_salary_scratch = model.intercept_ + model.coef_[0] * experience + model.coef_[1] * education
print(f"Predicted Salary  for 3 years of Experience and Education level 5: ${predicted_salary_scratch:.2f}")

# Create a 3D plot for the regression surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual data points
ax.scatter(df['Experience'], df['Education'], df['Salary'], color='blue', label='Actual Data')

# Create a grid of points for plotting the plane
exp_range = np.linspace(df['Experience'].min(), df['Experience'].max(), 10)
edu_range = np.linspace(df['Education'].min(), df['Education'].max(), 10)
exp_grid, edu_grid = np.meshgrid(exp_range, edu_range)
salary_grid = model.intercept_ + model.coef_[0] * exp_grid + model.coef_[1] * edu_grid
# Plot the regression plane
ax.plot_surface(exp_grid, edu_grid, salary_grid, color='red', alpha=0.5)
# Labels and title
ax.set_xlabel('Experience')
ax.set_ylabel('Education')
ax.set_zlabel('Salary')
ax.set_title('3D Linear Regression: Experience, Education vs Salary')

# Show the plot
plt.show()
```
