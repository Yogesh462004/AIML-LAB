## SIMPLE LINEAR REGRESSION

<H3>Manual</H3>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("data1.csv")
x = data["YearsExperience"]
y = data["Salary"]

# Calculate means
x_mean = x.mean()
y_mean = y.mean()

# Calculate slope and intercept
numerator = ((x - x_mean) * (y - y_mean)).sum()
denominator = ((x - x_mean) ** 2).sum()
slope = numerator / denominator
intercept = y_mean - slope * x_mean

# Predict y values
y_pred = slope * x + intercept

# Calculate R^2
ss_total = ((y - y_mean) ** 2).sum()
ss_residual = ((y - y_pred) ** 2).sum()
r_squared = 1 - (ss_residual / ss_total)

# Plot
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', label=f'Best Fit: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Output results
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R^2: {r_squared:.2f}")

# Predict based on user input
x_input = float(input("\nEnter years of experience to predict salary: "))
y_output = slope * x_input + intercept
print(f"Predicted Salary for {x_input} years of experience: {y_output:.2f}")
```

<h3>Library</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("data1.csv")
x = data[["YearsExperience"]]  # 2D input with column name
y = data["Salary"]

# Create and train the model
model = LinearRegression()
model.fit(x, y)

# Get slope, intercept, and R^2
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(x, y)

# Predict y values for plotting
y_pred = model.predict(x)

# Plot data and regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', label=f'Best Fit: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Simple Linear Regression with scikit-learn')
plt.legend()
plt.grid(True)
plt.show()

# Print results with  decimal places
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R^2 Score: {r_squared:.4f}")

# Predict for user input
x_input = float(input("\nEnter years of experience to predict salary: "))
x_input_df = pd.DataFrame([[x_input]], columns=["YearsExperience"])
y_output = model.predict(x_input_df)
print(f"Predicted Salary for {x_input:.2f} years of experience: {y_output[0]:.2f}")
```
