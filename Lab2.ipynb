```python
import matplotlib.pyplot as plt

def draw_board(ax, board, move_num, player):
    """Draws a single Tic-Tac-Toe board in a subplot."""
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.set_title(f"{move_num} - {player} Turn")
    
    # Draw grid lines
    for i in range(1, 3):
        ax.plot([0, 3], [i, i], color="black", linewidth=2)
        ax.plot([i, i], [0, 3], color="black", linewidth=2)

    # Plot Xs and Os
    for i in range(3):
        for j in range(3):
            if board[i][j] == "X":
                ax.text(j + 0.5, 2.5 - i, "X", fontsize=20, ha="center", va="center", color="blue")
            elif board[i][j] == "O":
                ax.text(j + 0.5, 2.5 - i, "O", fontsize=20, ha="center", va="center", color="black")

def visualize_game(moves):
    """Creates a visualization of a Tic-Tac-Toe game progression."""
    fig, axes = plt.subplots(2, 4, figsize=(10, 5))  # Adjusted for 8 turns (2 rows x 4 columns)
    axes = axes.flatten()
    
    board = [[" " for _ in range(3)] for _ in range(3)]
    
    for move_num, (row, col, player) in enumerate(moves, start=1):
        board[row][col] = player
        draw_board(axes[move_num - 1], board, move_num, player)

    # Display result
    axes[-1].set_title("X Wins", fontsize=14, fontweight="bold")  # Customize based on result
    axes[-1].axis("off")  # Hide last empty subplot
    
    plt.tight_layout()
    plt.show()

# Example game moves (row, col, player)
game_moves = [
    (0, 0, "X"), (1, 1, "O"), (0, 1, "X"), (2, 2, "O"),
    (1, 0, "X"), (1, 2, "O"), (2, 0, "X")
]

# Generate visualization
visualize_game(game_moves)
```

```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board, player):
    # Check rows and columns
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or all(board[j][i] == player for j in range(3)):
            return True
    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

def get_available_moves(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == " "]

def machine_move(board):
    return random.choice(get_available_moves(board))

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    user_symbol = "X"
    machine_symbol = "O"
    
    # Ask who should go first
    while True:
        first_player = input("Do you want to go first? (yes/no): ").strip().lower()
        if first_player in ["yes", "no"]:
            break
        print("Invalid input! Please enter 'yes' or 'no'.")

    user_turn = first_player == "yes"
    
    print("Welcome to Tic-Tac-Toe! You are 'X', and the machine is 'O'.")
    print_board(board)
    
    for turn in range(9):
        if user_turn:
            # User's turn
            while True:
                try:
                    row, col = map(int, input("Enter your move (row and column, space-separated): ").split())
                    if (row, col) in get_available_moves(board):
                        board[row][col] = user_symbol
                        break
                    else:
                        print("Invalid move! Try again.")
                except ValueError:
                    print("Please enter two numbers between 0 and 2.")
        else:
            # Machine's turn
            row, col = machine_move(board)
            board[row][col] = machine_symbol
            print(f"Machine placed 'O' at ({row}, {col})")
        
        print_board(board)
        
        # Check for a winner
        if check_winner(board, user_symbol):
            print("Congratulations! You win!")
            return
        if check_winner(board, machine_symbol):
            print("Machine wins! Better luck next time.")
            return
        
        # Switch turns
        user_turn = not user_turn
    
    print("It's a draw!")

tic_tac_toe()
```
